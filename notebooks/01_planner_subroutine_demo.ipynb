## üß† Planner inputs

```python
model = os.getenv("FACTICLI_MODEL", "gpt-4.1-mini")

claim = "The Eiffel Tower was built in 1889 for the World's Fair."
# claim = "The Great Wall of China is visible from space with the naked eye."
# claim = "The first iPhone was released in 2007."
# claim = "Coffee was discovered in Brazil in the 18th century."

planner_prompt = (
    "Build a fact-checking plan for this claim.\\n\\n"
    f"Claim:\\n{claim}\\n\\n"
    "Output at most 6 checks."
)

print(model)
print(claim)
```

## üöÄ Run planner agent

```python
from agents import Runner
from facticli.agents import build_planner_agent
from facticli.types import InvestigationPlan

planner = build_planner_agent(model=model)
result = await Runner.run(planner, planner_prompt)
plan = result.final_output_as(InvestigationPlan, raise_if_incorrect_type=True)

plan.model_dump()
```
